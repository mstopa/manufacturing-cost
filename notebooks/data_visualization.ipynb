## Fetch numeric data

```python
df = pd.read_csv('../data/train_numeric.csv')
```

```python
df_true = df[df['Response'] == 1]
df_true
```

```python
df_false = df[df['Response'] == 0]
df_false
```

```python
df = pd.concat([df_true.sample(6000), df_false.sample(6000)])
df
```

## Clean and normalize data

```python
# Calculate percentage of None values for each column
null_percentages = df.isna().mean()

# Set a threshold for percentage of None values
threshold = 0.8

# Filter for columns where percentage of None values exceeds the threshold
cols_to_drop = null_percentages[null_percentages > threshold].index

# Drop the columns from the dataframe
df.drop(cols_to_drop, axis=1, inplace=True)

df
```

```python
# Normalize the column values and replace NaN values with normalized column values for all columns
df.fillna(df.mean(), inplace=True)
df
```

```python
from sklearn.preprocessing import StandardScaler

# Select all columns except for 'Id'
cols_to_normalize = [col for col in df.columns if col not in ('Id', 'Response')]

normalized_df = pd.DataFrame(StandardScaler().fit_transform(df[cols_to_normalize]), columns = cols_to_normalize, index=df.index)

# Add the 'Id' column back to the normalized dataframe
normalized_df.insert(0, 'Id', df['Id'])
normalized_df.insert(1, 'Response', df['Response'])

normalized_df
```

## Remove highly correlating features

```python
cor_matrix = normalized_df[cols_to_normalize].corr().abs()
print(cor_matrix)
```

```python
import numpy as np

upper_tri = cor_matrix.where(np.triu(np.ones(cor_matrix.shape),k=1).astype(bool))
upper_tri
```

```python
to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > 0.95)]
print(to_drop)
len(to_drop)
```

```python
df_filtered = normalized_df.drop(normalized_df[to_drop], axis=1)
df_filtered
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x='L0_S0_F0', y='L0_S0_F2', hue='Response', data=df_filtered)
plt.show()
```

```python
class_0 = df_filtered[df_filtered['Response'] == 0]
class_1 = df_filtered[df_filtered['Response'] == 1]

# Create histograms for each feature
features = df_filtered.columns[2:]
features = np.random.choice(features, 5)

for feature in features:
    plt.hist(class_0[feature], alpha=0.5, label='OK')
    plt.hist(class_1[feature], alpha=0.5, label='Failure')
    plt.title(f'Histogram of {feature}')
    plt.legend()
    plt.show()
```

## Split data

```python
from sklearn.model_selection import train_test_split

def split(_df):
    # Split the data into training and testing sets (80/20 split)
    train_df, test_df = train_test_split(_df, test_size=0.2, random_state=42)

    # Split the training data into training and validation sets (80/20 split)
    train_df, dev_df = train_test_split(train_df, test_size=0.2, random_state=42)

    train_df["subset"] = "train"
    dev_df["subset"] = "dev"
    test_df["subset"] = "test"

    return pd.concat([train_df, dev_df, test_df])

df_1 = df_filtered[df_filtered['Response'] == 1]
df_0 = df_filtered[df_filtered['Response'] == 0]

df_1 = split(df_1)
df_0 = split(df_0)

df = pd.concat([df_1, df_0])
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

train_0 = df[(df["subset"]=="train") & (df["Response"]==0)].shape[0]
train_1 = df[(df["subset"]=="train") & (df["Response"]==1)].shape[0]
test_0 = df[(df["subset"]=="test") & (df["Response"]==0)].shape[0]
test_1 = df[(df["subset"]=="test") & (df["Response"]==1)].shape[0]
dev_0 = df[(df["subset"]=="dev") & (df["Response"]==0)].shape[0]
dev_1 = df[(df["subset"]=="dev") & (df["Response"]==1)].shape[0]

data = {
    "Subset": ["train", "train", "test", "test", "dev", "dev"],
    "Response": [0, 1] * 3,
    "Frequency": [train_0, train_1, test_0, test_1, dev_0, dev_1]
}

sns.histplot(data=data, x="Response", hue="Subset", weights="Frequency", multiple="stack", bins=4)
plt.xticks([0, 1], ["0", "1"])
plt.xlabel("Response")
plt.ylabel("Frequency")
plt.show()
```

```python
df.to_csv('../data/processed_data.csv', index=False)
```

```python
X_train = df[df["subset"]=="train"].drop(["Id", "Response", "subset"], axis=1)
y_train = df[df["subset"]=="train"]["Response"]
X_test = df[df["subset"]=="test"].drop(["Id", "Response", "subset"], axis=1)
y_test = df[df["subset"]=="test"]["Response"]
```

## Linear regression

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, predictions)

TN, FP, FN, TP = confusion_matrix(y_test, predictions).ravel()

print('True Positive(TP)  = ', TP)
print('False Positive(FP) = ', FP)
print('True Negative(TN)  = ', TN)
print('False Negative(FN) = ', FN)

accuracy =  (TP+TN) /(TP+FP+TN+FN)

print('Accuracy of the binary classification = {:0.3f}'.format(accuracy))

sns.set(font_scale=1.2)
sns.heatmap(cm, annot=True, fmt='g', cmap='Blues', annot_kws={'size': 16}, cbar=False)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
models = {}

# Logistic Regression
from sklearn.linear_model import LogisticRegression
models['Logistic Regression'] = LogisticRegression()

# Support Vector Machines
from sklearn.svm import LinearSVC
models['Support Vector Machines'] = LinearSVC()

# Decision Trees
from sklearn.tree import DecisionTreeClassifier
models['Decision Trees'] = DecisionTreeClassifier()

# Random Forest
from sklearn.ensemble import RandomForestClassifier
models['Random Forest'] = RandomForestClassifier()

# Naive Bayes
from sklearn.naive_bayes import GaussianNB
models['Naive Bayes'] = GaussianNB()

# K-Nearest Neighbors
from sklearn.neighbors import KNeighborsClassifier
models['K-Nearest Neighbor'] = KNeighborsClassifier()
from sklearn.metrics import accuracy_score, precision_score, recall_score

accuracy, precision, recall = {}, {}, {}

for key in models.keys():
    
    # Fit the classifier model
    models[key].fit(X_train, y_train)
    
    # Prediction 
    predictions = models[key].predict(X_test)
    
    # Calculate Accuracy, Precision and Recall Metrics
    accuracy[key] = accuracy_score(predictions, y_test)
    precision[key] = precision_score(predictions, y_test)
    recall[key] = recall_score(predictions, y_test)
```

```python
# Neural Networks
from keras.models import Sequential
from keras.layers import Dense

model = Sequential() 
model.add(Dense(128, activation='relu', input_dim=164))
model.add(Dense(1, activation='sigmoid')) 
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy']) 
model.summary()
```

```python
X_dev = df[df["subset"]=="dev"].drop(["Id", "Response", "subset"], axis=1)
y_dev = df[df["subset"]=="dev"]["Response"]

hist = model.fit(X_train, y_train, validation_data=(X_dev, y_dev), epochs=10, batch_size=100)
```

```python
nn_preds = model.predict(X_test)

threshold = 0.5

nn_preds = [1 if x > threshold else 0 for x in nn_preds]
models["NN"] = model
accuracy["NN"] = accuracy_score(nn_preds, y_test)
precision["NN"] = precision_score(nn_preds, y_test)
recall["NN"] = recall_score(nn_preds, y_test)
```

```python
nn_preds
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

sns.set()
acc = hist.history['accuracy']
val = hist.history['val_accuracy']
epochs = range(1, len(acc) + 1)

plt.plot(epochs, acc, '-', label='Training accuracy')
plt.plot(epochs, val, ':', label='Validation accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
plt.plot()
```

```python
import pandas as pd

df_model = pd.DataFrame(index=models.keys(), columns=['Accuracy', 'Precision', 'Recall'])

round_list = lambda x: [round(i, 2) for i in x]

df_model['Accuracy'] = round_list(accuracy.values())
df_model['Precision'] = round_list(precision.values())
df_model['Recall'] = round_list(recall.values())

df_model
```

```python
ax  = df_model.plot.bar(rot=45)
ax.legend(ncol= len(models.keys()), bbox_to_anchor=(0, 1), loc='lower left', prop={'size': 14})
plt.tight_layout()
```
